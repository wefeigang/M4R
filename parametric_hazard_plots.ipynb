```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# x-axis
time_range = np.linspace(0.01, 10, 500)

lambda_list = [0.01, 0.1, 1.0]

# exponetial
plt.figure(figsize=(10, 8))

for lam in lambda_list:
    # Exponential hazard function
    hazard = lam * np.ones_like(time_range)

    # Plotting
    plt.plot(time_range, hazard, label=rf'$\lambda$={lam}')

plt.xlabel('Time')
plt.ylabel('Hazard function')
plt.legend()
plt.show()
```

```python
# pdf plot
plt.figure(figsize=(10, 8))
for lam in lambda_list:
    # Exponential PDF
    pdf = lam * np.exp(-lam * time_range)

    # Plotting
    plt.plot(time_range, pdf, label=rf'$\lambda$={lam}')
plt.xlabel('Time')
plt.ylabel('Probability density function')
plt.legend()
```

```python
# weibull
time_range = np.linspace(0.01, 50, 500)  # new range

gamma_list = [0.5, 1.5, 3]  # shape parameters for Weibull
median_time = 20  # keep median constant

plt.figure(figsize=(10, 8))
for gamma in gamma_list:
    # Weibull hazard function
    lam = np.log(2) / (median_time ** gamma)  # compute lambda
    hazard = lam * gamma * (time_range ** (gamma - 1))

    # Plotting
    plt.plot(time_range, hazard, label=rf'$\gamma$={gamma}')

plt.xlabel('Time')
plt.ylabel('Hazard function')
plt.ylim(-0.005, 0.2)
plt.yticks(np.arange(0, 0.21, 0.05))
plt.legend()
plt.show()
```
