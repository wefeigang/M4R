```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# x-axis
time_range = np.linspace(0.01, 10, 500)

lambda_list = [0.01, 0.1, 1.0]

# exponetial
plt.figure(figsize=(10, 8))

for lam in lambda_list:
    # Exponential hazard function
    hazard = lam * np.ones_like(time_range)

    # Plotting
    plt.plot(time_range, hazard, label=rf'$\lambda$={lam}')

plt.xlabel('Time')
plt.ylabel('Hazard function')
plt.legend()
plt.show()
```

```python
# pdf plot
plt.figure(figsize=(10, 8))
for lam in lambda_list:
    # Exponential PDF
    pdf = lam * np.exp(-lam * time_range)

    # Plotting
    plt.plot(time_range, pdf, label=rf'$\lambda$={lam}')
plt.xlabel('Time')
plt.ylabel('Probability density function')
plt.legend()
```

```python
# weibull
gamma_list = [0.5, 1.0, 1.5]  # shape parameters for Weibull

plt.figure(figsize=(10, 8))
for gamma in gamma_list:
    # Weibull hazard function with shape parameter k=1.5
    lam = 1.5
    hazard = lam * gamma * (time_range ** (gamma - 1))

    # Plotting
    plt.plot(time_range, hazard, label=rf'$\gamma$={gamma}')

plt.xlabel('Time')
plt.ylabel('Hazard function')
plt.legend()
plt.show()
```
